# Task 1
Write Python code to develop a word guessing game where employees attempt to guess a secret word from a provided list by making a series of guesses. Clues, such as optional information about the number of letters or characters in the word, can be given to assist in the guessing process


```python
import random

def play_game():
    words = ['almond','butter', 'cherry', 'orange', 'peanut', 'tomato']
    secret_word = random.choice(words)
    guesses = set()
    attempts = 6

    print(f"Guess the word! It has {len(secret_word)} letters.")

    while attempts > 0:
        display = ' '.join([char if char in guesses else '_' for char in secret_word])
        print(f"\nWord: {display}")
        print(f"Attempts left: {attempts}")

        guess = input("Guess a letter: ").strip().lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Invalid input. Enter a single letter.")
            continue

        if guess in guesses:
            print("Already guessed.")
            continue

        guesses.add(guess)

        if guess in secret_word:
            print("Correct!")
        else:
            print("Wrong!")
            attempts -= 1

        if all(char in guesses for char in secret_word):
            print(f"Congratulations! The word was: {secret_word}")
            return

    print(f"Game over! The word was: {secret_word}")

if __name__ == "__main__":
    play_game()
```

# Task 2
Define a method named menu() that prints the following options of the application

1. Play Riddle Puzzle

2. Play Arithmetic Puzzle

3. Play General Knowledge quiz game.

4. Play Rock-Paper-Scissor game with the computer

5. Play Rock-Paper-Scissor game with a user

6. Play Word-Guessing game with the computer

Enter any option by typing a number between 1 and 6.

Finally, write the main() method and observe the program's behavior as follows


1. The program will start by printing the application menu.

2. The program will prompt the user to enter an option by choosing a number between 1 and 6.

3. If the user enters the option value as 1, then execute the program to play the riddle puzzle

4. If the user enters the option value as 2, then execute the program to play the arithmetic puzzle.

5. If the user enters the option value as 3, then executethe program to play the General Knowledge quiz game

6. If the user enters the option value as 4, then execute the program to play the rock-paper-scissors game with the computer

7. If the user enters the option value as 5. then execute the program to play the rock-paper-scissors game between users
8. If the user enters the option value as 6, then execute the program to play word guessing game.

9. If the user enters any value outside the range of 1 to 6, then the program will display You have entered the wrong option

10. The program will prompt the user to enter their choice (Y/y for Yes or N/n for No) to either repeat the menu to play another game or quit the main program


Ensure to include exception handling to manage possible errors during the execution of the games Additionally, specify that the menu() method should be defined as a static method within the GameMenu class to allow direct access using the class name.:

```python
import random

# Riddle Puzzle
def play_riddle_puzzle():
    riddles = {
        "What can run but cannot walk?": "water",
        "What comes down but never goes up?": "rain",
        "What has one eye but can't see?": "needle",
        "What starts with T, end with T and has T inside it?": "teapot",
        "I have no wings, but I can fly. And I have no eyes, but I can cry. what am I?": "cloud"
    }
    correct = total = 0

    while True:
        question, answer = random.choice(list(riddles.items()))
        user_answer = input(f"\nRiddle: {question}\nYour answer: ").strip().lower()

        if user_answer == answer:
            print("Correct!")
            correct += 1
        else:
            print(f"Incorrect. The correct answer is '{answer}'.")

        total += 1
        play_again = input("Play again? (yes/no): ").strip().lower()

        if play_again != 'yes':
            break

    print(f"\nYour score: {correct}/{total} ({(correct / total) * 100:.2f}%)")

# Arithmetic Puzzle
def play_arithmetic_puzzle():
    correct = total = 0

    while True:
        num1, num2 = random.randint(1, 20), random.randint(1, 20)
        operator = random.choice(['+', '-', '*', '/'])

        if operator == '+':
            correct_answer = num1 + num2
        elif operator == '-':
            correct_answer = num1 - num2
        elif operator == '*':
            correct_answer = num1 * num2
        else:
            correct_answer = round(num1 / num2, 2)

        print(f"\nSolve: {num1} {operator} {num2}")

        try:
            user_answer = input("Your answer: ").strip()
            if user_answer.lower() == 'exit':
                break
            if float(user_answer) == correct_answer:
                print("Correct!")
                correct += 1
            else:
                print(f"Incorrect! The correct answer is {correct_answer}")
        except ValueError:
            print("Invalid input! Please enter a number.")

        total += 1
        if input("Play again? (yes/no): ").strip().lower() != 'yes':
            break

    print(f"\nYour score: {correct}/{total} ({(correct / total) * 100:.2f}%)")

# General Knowledge Quiz
def play_gk_quiz_game():
    science_questions = [
        {"question": "Which gas do plants absorb during photosynthesis?", "answer": "carbon dioxide"},
        {"question": "What is the full name of the WWW?", "answer": "world wide web"},
        {"question": "How many bones are there in the adult human body?", "answer": "206"}
    ]

    math_questions = [
        {"question": "What is the only even prime number?", "answer": "2"},
        {"question": "What is the sum of the first 5 natural numbers?", "answer": "15"},
        {"question": "What is the name of the mathematical symbol Ï€?", "answer": "pi"}
    ]

    geography_questions = [
        {"question": "What is the tallest mountain on Earth?", "answer": "mount everest"},
        {"question": "What is the capital of India?", "answer": "delhi"},
        {"question": "Which planet is known as the 'Red Planet'?", "answer": "mars"}
    ]

    other_questions = [
        {"question": "Which country won the most recent FIFA World Cup (as of 2022)?", "answer": "argentina"},
        {"question": "Who invented the light bulb?", "answer": "thomas edison"},
        {"question": "In what year did India declare itself a Democratic and Republic state?", "answer": "1950"}
    ]

    categories = {
        "1": science_questions,
        "2": math_questions,
        "3": geography_questions,
        "4": other_questions
    }

    def ask_category():
        print("\nSelect a category:")
        print("1. Science\n2. Mathematics\n3. Geography\n4. Others")

        while True:
            category = input("Enter the category number (1-4) or type 'exit' to quit: ").strip()
            if category.lower() == "exit":
                return None
            if category in categories:
                return category
            else:
                print("Invalid category selection. Please choose a number between 1 and 4.")

    def ask_question(question_data):
        return input(f"\n{question_data['question']}\nYour answer: ").strip().lower()

    def check_answer(user_answer, correct_answer):
        if user_answer == correct_answer:
            print("Correct!")
            return True
        else:
            print(f"Incorrect! The correct answer is: {correct_answer}")
            return False

    def exec_questions(category):
        correct = total = 0
        for _ in range(len(categories[category])):
            question_data = random.choice(categories[category])
            user_answer = ask_question(question_data)
            if check_answer(user_answer, question_data["answer"]):
                correct += 1
            total += 1
        return correct, total

    def display_score(correct, total):
        if total > 0:
            percentage = (correct / total) * 100
            print(f"\nYour final score: {correct}/{total} ({percentage:.2f}%)")
        else:
            print("No questions were answered.")

    print("Welcome to the General Knowledge Quiz Game!")
    total_correct = total_questions = 0

    while True:
        category = ask_category()
        if category is None:
            break

        correct, total = exec_questions(category)
        total_correct += correct
        total_questions += total

        if input("Do you want to play another category? (yes/no): ").strip().lower() != 'yes':
            break

    display_score(total_correct, total_questions)

# Rock-Paper-Scissors with Computer
def play_rps_computer():
    choices = ['rock', 'paper', 'scissors']

    def determine_winner(player_choice, computer_choice):
        if player_choice == computer_choice:
            return "It's a tie!"
        elif (player_choice == 'rock' and computer_choice == 'scissors') or \
             (player_choice == 'scissors' and computer_choice == 'paper') or \
             (player_choice == 'paper' and computer_choice == 'rock'):
            return "You win!"
        else:
            return "Computer wins!"

    while True:
        computer_choice = random.choice(choices)
        player_choice = input("\nEnter your choice (rock, paper, scissors) or type 'exit' to quit: ").lower()

        if player_choice == 'exit':
            print("Thanks for playing!")
            break

        if player_choice not in choices:
            print("Invalid choice! Please choose rock, paper, or scissors.")
            continue

        print(f"\nComputer chose: {computer_choice}")
        print(f"You chose: {player_choice}")

        result = determine_winner(player_choice, computer_choice)
        print(result)

# Rock-Paper-Scissors between Users
def play_rps_user():
    def determine_winner(player1_choice, player2_choice, player1_name, player2_name):
        if player1_choice == player2_choice:
            return "It's a tie!"
        elif (player1_choice == 'rock' and player2_choice == 'scissors') or \
             (player1_choice == 'scissors' and player2_choice == 'paper') or \
             (player1_choice == 'paper' and player2_choice == 'rock'):
            return f"{player1_name} wins!"
        else:
            return f"{player2_name} wins!"

    print("Welcome to the two-player Rock-Paper-Scissors game!")
    player1_name = input("Hello Player 1! Enter your name: ")
    player2_name = input("Hello Player 2! Enter your name: ")

    player1_choice = input(f"\n{player1_name}, choose rock, paper, or scissors: ").lower()
    player2_choice = input(f"\n{player2_name}, choose rock, paper, or scissors: ").lower()

    if player1_choice not in ['rock', 'paper', 'scissors'] or player2_choice not in ['rock', 'paper', 'scissors']:
        print("Invalid choice! Please enter rock, paper, or scissors.")
        return

    result = determine_winner(player1_choice, player2_choice, player1_name, player2_name)
    print(result)

# Word Guessing Game with Computer
def play_word_guessing_game():
    words = ['almond', 'butter', 'cherry', 'orange', 'peanut', 'tomato']
    secret_word = random.choice(words)
    guesses = set()
    attempts = 6

    print(f"Guess the word! It has {len(secret_word)} letters.")

    while attempts > 0:
        display = ' '.join([char if char in guesses else '_' for char in secret_word])
        print(f"\nWord: {display}")
        print(f"Attempts left: {attempts}")

        guess = input("Guess a letter: ").strip().lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Invalid input. Enter a single letter.")
            continue

        if guess in guesses:
            print(f"You already guessed '{guess}'!")
            continue

        guesses.add(guess)

        if guess in secret_word:
            print(f"Good guess! '{guess}' is in the word.")
            if all(char in guesses for char in secret_word):
                print(f"Congratulations! You guessed the word: {secret_word}")
                break
        else:
            attempts -= 1
            print(f"Wrong guess! '{guess}' is not in the word.")

    if attempts == 0:
        print(f"Game over! The word was '{secret_word}'.")

# Game Menu
class GameMenu:
    @staticmethod
    def menu():
        while True:
            print("\nSelect a game to play:")
            print("1. Riddle Puzzle")
            print("2. Arithmetic Puzzle")
            print("3. General Knowledge Quiz")
            print("4. Rock-Paper-Scissors (Computer)")
            print("5. Rock-Paper-Scissors (2 Players)")
            print("6. Word Guessing Game")
            print("7. Quit")

            try:
                choice = int(input("Enter your choice: ").strip())
                if choice == 1:
                    play_riddle_puzzle()
                elif choice == 2:
                    play_arithmetic_puzzle()
                elif choice == 3:
                    play_gk_quiz_game()
                elif choice == 4:
                    play_rps_computer()
                elif choice == 5:
                    play_rps_user()
                elif choice == 6:
                    play_word_guessing_game()
                elif choice == 7:
                    break
                    print("Exiting game menu. Goodbye!")

                else:
                    print("Invalid choice! Please select a number between 1 and 7.")
            except ValueError:
                print("Invalid input! Please enter a number.")

if __name__ == "__main__":
    GameMenu.menu()
```
