# Task 1

Write a program to develop a simple console-based computer brain teaser, a riddle logic puzzle, text-based puzzles involving wordplay, logic, and lateral thinking games in Python. The program will randomly choose a riddle question from a list of riddles (Que - Ans dictionary) provided in program. It will then prompt the user for an answer and compare it against the correct answer.

If the user's answer matches the correct one, the program will print 'Correct'. else, it will print 'Incorrect' and provide correct answer. The program will continue to run infinitely, allowing users to solve multiple riddles with the option to play again. Upon exiting the game, the program should display the user's score as a percentage


```python
import random
                # Riddles stored as question-answer pairs in a dictionary
riddles = {
    "What can run but cannot walk?": "water",
    "What comes down but never goes up?": "rain",
    "What has one eye but can't see?": "needle",
    "What starts with T, end with T and has T inside it?": "teapot",
    "I have no wings, but i can fly. And I have no eyes, but I can cry. what am I?": "cloud"
}
correct = total = 0

while True:
    question, answer = random.choice(list(riddles.items()))
    user_answer = input(f"\nRiddle: {question}\nYour answer: ").strip().lower()

    if user_answer == answer:
        print("Correct!")
        correct += 1
    else:
        print(f"Incorrect. The correct answer is '{answer}'.")

    total += 1
    play_again = input("Play again? (yes/no): ").strip().lower()

    if play_again != 'yes':
        break

print(f"\nYour score: {correct}/{total} ({(correct / total) * 100:.2f}%)")
```

# Task 2

Write a program to develop a simple console-based computer brain teaser, an arithmetic logic puzzle, a problem-solving game in Python. The program will randomly generate two numeric numbers within the range of 1 to 20 and an arithmetic operator ( +, -, /, * )The program will run infinite turns to solve the arithmetic puzzle with the option to play again. 13

The program should handle exceptions gracefully, such as invalid inputs when user attempts to solve the puzzle The program should print the percentage of score when user leaves the game.


```python
import random
correct = total = 0

while True:
    num1, num2 = random.randint(1, 20), random.randint(1, 20)
    operator = random.choice(['+', '-', '*', '/'])

    if operator == '+':
        correct_answer = num1 + num2
    elif operator == '-':
        correct_answer = num1 - num2
    elif operator == '*':
        correct_answer = num1 * num2
    else:
        correct_answer = round(num1 / num2, 2)

    print(f"\nSolve: {num1} {operator} {num2}")

    try:
        user_answer = input("Your answer: ").strip()
        if user_answer.lower() == 'exit':
            break
        if float(user_answer) == correct_answer:
            print("Correct!")
            correct += 1
        else:
            print(f"Incorrect! The correct answer is {correct_answer}")
    except ValueError:
        print("Invalid input! Please enter a number.")

    total += 1
    if input("Play again? (yes/no): ").strip().lower() != 'yes':
        break

if total > 0:
    print(f"\nYour score: {correct}/{total} ({(correct / total) * 100:.2f}%)")
else:
    print("No puzzles were answered.")
```

# Task 3
Write a Python program to develop a General Knowledge quiz game Create four lists to store general knowledge questions of four categories: Science, Mathematics, Geography and others. Each item in these lists is a dictionary with "question" and "answer" keys, as provided in the upcoming slide. Ensure to include proper exception handling to handle invalid user inputs
The program should display questions based on the category selected by the user. After the user submits an answer, the program should indicate whether it is correct. If the answer provided by the user is incorrect, the program should display the correct answer List of questions and answers:

Create the following methods for game logic:
ask_category(): Presents category options for the user to choose from.
ask_question(question): Displays a given question and prompts the user for an answer. check_answer(user_answer, correct_answer): Compares the user's answer with the correct answer and provides feedback. display_score(questions, score) Prints a summary of the player's performance, including correct answers, total questions, and percentage of correct responses exec_questions(questions) Iterates through a given set of questions, handles user interactions, and keeps track of the score play_game(): Manages the main game loop, including category selection, question execution, and the option to play again. Incorporate exception handling to ensure the program gracefully handles invalid inputs and unexpected errors

Run the Program:
The if name == main": block ensures code execution only when the program is run directly.
Prints a welcome message.
Calls the play_game() method to start the game.
Prints a thank you message after the game concludes.
Here is an example of how the program should work:

Choose your category from the following list:
1. Science-based
2. Mathematics-based
3. Geography-based
4. Others

Suppose 2nd category is chosen by user, then a Mathematics based question should be displayed, such as 'What is the only even prime number? The user will provide their answer If the answer is correct, the program should
display 'Correct' If the answer is incorrect, the program should display: 'Incorrect. The correct answer is 2'



```python
import random
                        # Questions for each category
science_questions = [
    {"question": "Which gas do plants absorb during photosynthesis?", "answer": "carbon dioxide"},
    {"question": "What is the full name of the WWW?", "answer": "world wide web"},
    {"question": "How many bones are there in the adult human body?", "answer": "206"}
]

math_questions = [
    {"question": "What is the only even prime number?", "answer": "2"},
    {"question": "What is the sum of the first 5 natural numbers?", "answer": "15"},
    {"question": "What is the name of the mathematical symbol Ï€?", "answer": "pi"}
]

geography_questions = [
    {"question": "What is the tallest mountain on Earth?", "answer": "mount everest"},
    {"question": "What is the capital of India?", "answer": "delhi"},
    {"question": "Which planet is known as the 'Red Planet'?", "answer": "mars"}
]

other_questions = [
    {"question": "Which country won the most recent FIFA World Cup (as of 2022)?", "answer": "argentina"},
    {"question": "Who invented the light bulb?", "answer": "thomas edison"},
    {"question": "In what year did India declare itself a Democratic and Republic state?", "answer": "1950"}
]

categories = {      # Dictionary to categorize questions
    "1": science_questions,
    "2": math_questions,
    "3": geography_questions,
    "4": other_questions
}

def ask_category():      # Function to present category options to the user
    print("\nSelect a category:")
    print("1. Science\n2. Mathematics\n3. Geography\n4. Others")

    while True:
        category = input("Enter the category number (1-4) or type 'exit' to quit: ").strip()
        if category.lower() == "exit":
            return None
        if category in categories:
            return category
        else:
            print("Invalid category selection. Please choose a number between 1 and 4.")

def ask_question(question_data):      # Function to display a question and get the user's answer
    return input(f"\n{question_data['question']}\nYour answer: ").strip().lower()

def check_answer(user_answer, correct_answer):  # Function to check if the user's answer is correct
    if user_answer == correct_answer:
        print("Correct!")
        return True
    else:
        print(f"Incorrect! The correct answer is: {correct_answer}")
        return False

def display_score(correct, total):   # Function to print the final score
    if total > 0:
        percentage = (correct / total) * 100
        print(f"\nYour final score: {correct}/{total} ({percentage:.2f}%)")
    else:
        print("No questions were answered.")

def exec_questions(category):    # Function to execute the questions for a chosen category
    correct = total = 0
    for _ in range(len(categories[category])):
        question_data = random.choice(categories[category])
        user_answer = ask_question(question_data)
        if check_answer(user_answer, question_data["answer"]):
            correct += 1
        total += 1
    return correct, total

# Main game function to manage the game flow
def play_game():
    print("Welcome to the General Knowledge Quiz Game!")

    total_correct = total_questions = 0

    while True:
        category = ask_category()  # Get category selection from the user
        if category is None:
            break

        correct, total = exec_questions(category)
        total_correct += correct
        total_questions += total

        if input("Do you want to play another category? (yes/no): ").strip().lower() != 'yes':
            break

    display_score(total_correct, total_questions)
    print("Thanks for playing!")

if __name__ == "__main__": # Run the game
    play_game()
```

# Task 4
Think about how the rules of the Rock-Paper-Scissors game can be translated into Python code. Build the game using a Python program where Player 1 is the computer, making random choices.


```python
import random

choices = ['rock', 'paper', 'scissors']

def determine_winner(player_choice, computer_choice): # Function to determine the winner
    if player_choice == computer_choice:
        return "It's a tie!"
    elif (player_choice == 'rock' and computer_choice == 'scissors') or \
         (player_choice == 'scissors' and computer_choice == 'paper') or \
         (player_choice == 'paper' and computer_choice == 'rock'):
        return "You win!"
    else:
        return "Computer wins!"

def play_game():
    while True:
        # Computer makes a random choice
        computer_choice = random.choice(choices)

        # Ask the user for their choice
        player_choice = input("\nEnter your choice (rock, paper, scissors) or type 'exit' to quit: ").lower()

        if player_choice == 'exit':
            print("Thanks for playing!")
            break

        if player_choice not in choices:   # Validate the player's input
            print("Invalid choice! Please choose rock, paper, or scissors.")
            continue

        print(f"\nComputer chose: {computer_choice}")
        print(f"You chose: {player_choice}")

        result = determine_winner(player_choice, computer_choice)   # Determine and display the winner
        print(result)

if __name__ == "__main__": # Run the game
    play_game()
```

# Task 5
Develop a Python program to simulate a two-player Rock- Paper-Scissors game. The program should prompt each player to enter their name, allow each player to choose rock, paper, or scissors, implement logic to determine the winner, and display the winner's name

Example:
Hello user 1! Enter your name: John
Hello user2! Enter your name: Eva
Hello user1! Enter a choice (rock, paper, scissors): rock
Hello user2! Enter a choice (rock, paper, scissors): paper
Paper covers rock! Eva win!

```python
# Function to determine the winner
def determine_winner(player1_choice, player2_choice, player1_name, player2_name):
    if player1_choice == player2_choice:
        return "It's a tie!"
    elif (player1_choice == 'rock' and player2_choice == 'scissors') or \
         (player1_choice == 'scissors' and player2_choice == 'paper') or \
         (player1_choice == 'paper' and player2_choice == 'rock'):
        return f"{player1_choice.capitalize()} beats {player2_choice}! {player1_name} wins!"
    else:
        return f"{player2_choice.capitalize()} beats {player1_choice}! {player2_name} wins!"

# Main function for the game
def play_game():
    print("Welcome to the two-player Rock-Paper-Scissors game!")

    # Get player names
    player1_name = input("Hello Player 1! Enter your name: ")
    player2_name = input("Hello Player 2! Enter your name: ")

    # Get choices from both players
    player1_choice = input(f"\nHello {player1_name}! Enter a choice (rock, paper, scissors): ").lower()
    player2_choice = input(f"Hello {player2_name}! Enter a choice (rock, paper, scissors): ").lower()

    # Validate inputs
    valid_choices = ['rock', 'paper', 'scissors']
    if player1_choice not in valid_choices or player2_choice not in valid_choices:
        print("Invalid choice! Please enter rock, paper, or scissors.")
        return

    # Determine the winner and display the result
    result = determine_winner(player1_choice, player2_choice, player1_name, player2_name)
    print(result)

# Run the game
if __name__ == "__main__":
    play_game()
```
